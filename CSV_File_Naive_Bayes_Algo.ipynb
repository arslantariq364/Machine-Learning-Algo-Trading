```python
import pandas as pd
df = pd.read_csv("student_data.csv")
data = df.values.tolist ()
```

```python
total_samples = len(data)

class_counts = {"Pass": 0, "Fail": 0}

for row in data:
    final_answer = row[-1]
    class_counts[final_answer] += 1

prior_yes = class_counts["Pass"] / total_samples
prior_no = class_counts["Fail"] / total_samples

print(f"Total Yes Days: {class_counts['Pass']} -> Prior Yes: {prior_yes:.3f}")
print(f"Total No Days: {class_counts['Fail']} -> Prior No: {prior_no:.3f}")
```

```python
feature_counts_yes = {}
feature_counts_no = {}

for row in data:
    final_answer = row[-1]

    for i in range(len(row) - 1):
        feature = row[i]

        if final_answer == "Pass":
            if feature not in feature_counts_yes:
                feature_counts_yes[feature] = 0
            feature_counts_yes[feature] += 1

        elif final_answer == "Fail":
            if feature not in feature_counts_no:
                feature_counts_no[feature] = 0
            feature_counts_no[feature] += 1

print("Features on Pass:", feature_counts_yes)
print("Features on Fail:", feature_counts_no)
```

```python
def predict_naive_bayes(new_st):
    score_yes = prior_yes
    score_no = prior_no

    for feature in new_st:
        # Multiply Yes score
        if feature in feature_counts_yes:
            prob = feature_counts_yes[feature] / class_counts["Pass"]
            score_yes = score_yes * prob
        else:
            score_yes = score_yes * 0.01

        # Multiply No score
        if feature in feature_counts_no:
            prob = feature_counts_no[feature] / class_counts["Fail"]
            score_no = score_no * prob
        else:
            score_no = score_no * 0.01

    print(f"Final Score for Pass: {score_yes:.5f}")
    print(f"Final Score for Fail: {score_no:.5f}")

    if score_yes > score_no:
        print("Prediction: Yes Pass!")
    else:
        print("Prediction: No, Fail")
```

```python
new_st=["High","High","Medium"]
predict_naive_bayes(new_st)
```

```python
for new_st in data:
    predict_naive_bayes(new_st)
```
