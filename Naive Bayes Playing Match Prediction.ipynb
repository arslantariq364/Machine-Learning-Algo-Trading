```python
 data = [
    ["Sunny", "Hot", "High", "Weak", "No"],
    ["Sunny", "Hot", "High", "Strong", "No"],
    ["Overcast", "Hot", "High", "Weak", "Yes"],
    ["Rain", "Mild", "High", "Weak", "Yes"],
    ["Rain", "Cool", "Normal", "Weak", "Yes"],
    ["Rain", "Cool", "Normal", "Strong", "No"],
    ["Overcast", "Cool", "Normal", "Strong", "Yes"],
    ["Sunny", "Mild", "High", "Weak", "No"],
    ["Sunny", "Cool", "Normal", "Weak", "Yes"],
    ["Rain", "Mild", "Normal", "Weak", "Yes"],
    ["Sunny", "Mild", "Normal", "Strong", "Yes"],
    ["Overcast", "Mild", "High", "Strong", "Yes"],
    ["Overcast", "Hot", "Normal", "Weak", "Yes"],
    ["Rain", "Mild", "High", "Strong", "No"]
]
```

```python
total_samples = len(data)

class_counts = {"Yes": 0, "No": 0}

for row in data:
    final_answer = row[-1]
    class_counts[final_answer] += 1

prior_yes = class_counts["Yes"] / total_samples
prior_no = class_counts["No"] / total_samples

print(f"Total Yes Days: {class_counts['Yes']} -> Prior Yes: {prior_yes:.3f}")
print(f"Total No Days: {class_counts['No']} -> Prior No: {prior_no:.3f}")
```

```python
feature_counts_yes = {}
feature_counts_no = {}

for row in data:
    final_answer = row[-1]

    for i in range(len(row) - 1):
        feature = row[i]

        if final_answer == "Yes":
            if feature not in feature_counts_yes:
                feature_counts_yes[feature] = 0
            feature_counts_yes[feature] += 1

        elif final_answer == "No":
            if feature not in feature_counts_no:
                feature_counts_no[feature] = 0
            feature_counts_no[feature] += 1

print("Features on Yes days:", feature_counts_yes)
print("Features on No days:", feature_counts_no)
```

```python
def predict_naive_bayes(new_day):
    score_yes = prior_yes
    score_no = prior_no

    for feature in new_day:
        # Multiply Yes score
        if feature in feature_counts_yes:
            prob = feature_counts_yes[feature] / class_counts["Yes"]
            score_yes = score_yes * prob
        else:
            score_yes = score_yes * 0.01

        # Multiply No score
        if feature in feature_counts_no:
            prob = feature_counts_no[feature] / class_counts["No"]
            score_no = score_no * prob
        else:
            score_no = score_no * 0.01

    print(f"Final Score for Yes: {score_yes:.5f}")
    print(f"Final Score for No: {score_no:.5f}")

    if score_yes > score_no:
        print("Prediction: Yes, play tennis!")
    else:
        print("Prediction: No, do not play tennis.")
```

```python
new_day=["Sunny","Cool","High","Strong","Dark"]
predict_naive_bayes(new_day)
```
